```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
x = np.linspace(0,10,100)
```

```python
y = np.sin(x)
```

```python
fig, ax = plt.subplots()
ax.plot(x ,y, label = "sin(x)",color = "pink")
ax.set_xlabel("x")
ax.set_ylabel("sin(x)")
ax.set_title("sine_wave")
ax.legend()
plt.show()
```

```python
np.random.seed(42)
epochs = np.arange(1,21)
train_loss = np.exp(-epochs/5)+np.random.rand(20)*0.5
```

```python
val_loss = np.exp(-epochs/6)+np.random.rand(20)*0.08
```

```python
plt.plot(epochs, train_loss,label = "train")
plt.plot(epochs, val_loss,label = "validity")
plt.xlabel("Epoch")
plt.ylabel("loss")
plt.title("Training vs validation loss")
plt.legend()
plt.show()
```

```python
#histogram
data = np.random.randn(1000)
```
